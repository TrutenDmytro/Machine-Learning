```python
# Трутень Дмитро, ФІТ, 4-4
# Варіант 28

import math
import random
import numpy as np

# Завдання 1
# Створіть одновимірний масив (вектор) зі значеннями n!, факторіала кожного натурального числа n, починаючи з 1, та виведіть його значення. 

def task_1():
    n = int(input("Введіть кількість елементів масиву (n): "))
    factorial_array = []

    for i in range(1, n + 1):
        factorial = math.factorial(i)
        factorial_array.append(factorial)

    print("Масив факторіалів:", factorial_array)

# Завдання 2
# Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його оберненими значеннями косинусів кожного натурального числа, починаючи з 1, та виведіть його значення.  

def task_2():
    # Розмірність матриці
    rows = 3
    cols = 3

    matrix = [[0] * cols for _ in range(rows)]

    for i in range(rows):
        for j in range(cols):
            n = i * cols + j + 1
            cos_value = math.cos(n)
            if cos_value != 0:
                inverse_cos = 1 / cos_value
                matrix[i][j] = inverse_cos

    # Виведення матриці
    for row in matrix:
        print(row)

# Завдання 3
# Створіть масив розміром 8x8, заповніть його випадковими дійсними числами від -1 до 1 та знайдіть відсоток чисел, які менше нуля та мають непарний індекс. 

def task_3():
    # Розмірність матриці
    rows = 8
    cols = 8

    matrix = [[random.uniform(-1, 1) for _ in range(cols)] for _ in range(rows)]

    # Лічильник для відслідковування чисел, які менше нуля та мають непарний індекс
    count_negative_odd_index = 0

    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] < 0 and (i + j) % 2 != 0:
                count_negative_odd_index += 1

    # Обчислення відсотка
    total_elements = rows * cols
    percentage = (count_negative_odd_index / total_elements) * 100

    # Виведення результату
    print("Масив:")
    for row in matrix:
        print(row)

    print(f"Відсоток чисел менше нуля з непарним індексом: {percentage:.2f}%")

# Завдання 4
# Створіть масив розміром 4x4, заповніть його випадковими дійсними числами від -1.4 до 1.4 та виведіть його значення. 

def task_4():
    # Розмірність матриці
    rows = 4
    cols = 4

    matrix = [[random.uniform(-1.4, 1.4) for _ in range(cols)] for _ in range(rows)]

    # Виведення матриці
    for row in matrix:
        print(row)

# Завдання 5
# Створіть два одновимірні масиви розміром 5, заповніть їх випадковими цілими числами в діапазоні від 2 до 7 та виконайте на них поелементні операції додавання, віднімання та множення.  

def task_5():
    # Розмірність масиву
    n = 5

    # Створення двох масивів та їх заповнення випадковими цілими числами
    array1 = [random.randint(2, 7) for _ in range(n)]
    array2 = [random.randint(2, 7) for _ in range(n)]

    print("Перший масив:", array1)
    print("Другий масив:", array2)

    # Поелементне додавання
    sum_array = [array1[i] + array2[i] for i in range(n)]
    print("Результат додавання:", sum_array)

    # Поелементне віднімання
    subtract_array = [array1[i] - array2[i] for i in range(n)]
    print("Результат віднімання:", subtract_array)

    # Поелементне множення
    multiply_array = [array1[i] * array2[i] for i in range(n)]
    print("Результат множення:", multiply_array)

# Завдання 6
# Створіть два вектори розміром 7, заповніть їх довільними цілими числами та знайдіть їх скалярний добуток. 

def task_6():
    vector1 = [random.randint(1, 10) for _ in range(7)]
    vector2 = [random.randint(1, 10) for _ in range(7)]

    scalar_product = sum(x * y for x, y in zip(vector1, vector2))
    print("Перший вектор:", vector1)
    print("Другий вектор:", vector2)
    print("Скалярний добуток:", scalar_product)

# Завдання 7
# Створіть дві матриці розміром 2x3 та 3x4, заповніть їх випадковими цілими числами від 1 до 15 та перемножте їх між собою. 

def task_7():
    # Розмірність першої матриці
    rows1 = 2
    cols1 = 3

    # Розмірність другої матриці
    rows2 = 3
    cols2 = 4

    matrix1 = [[random.randint(1, 15) for _ in range(cols1)] for _ in range(rows1)]
    matrix2 = [[random.randint(1, 15) for _ in range(cols2)] for _ in range(rows2)]

    # Перемноження матриць
    result_matrix = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(cols1)) for j in range(cols2)] for i in range(rows1)]

    # Виведення результату
    print("Перша матриця:")
    for row in matrix1:
        print(row)

    print("Друга матриця:")
    for row in matrix2:
        print(row)

    print("Результат множення:")
    for row in result_matrix:
        print(row)

# Завдання 8
# Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами від 1 до 15 та знайдіть її обернену матрицю.

def task_8():
    # Розмірність матриці
    rows = 3
    cols = 3

    matrix = [[random.randint(1, 15) for _ in range(cols)] for _ in range(rows)]

    try:
        # Знаходження оберненої матриці
        inverse_matrix = np.linalg.inv(matrix)

        # Виведення матриць
        print("Початкова матриця:")
        for row in matrix:
            print(row)

        print("Обернена матриця:")
        for row in inverse_matrix:
            print(row)

    except np.linalg.LinAlgError:
        print("Початкова матриця не має оберненої матриці.")

# Завдання 9
# Створіть матрицю розміром 7x7, заповніть її випадковими дійсними числами в діапазоні від 0 до 10 та транспонуйте її.

def task_9():
    # Розмірність матриці
    rows = 7
    cols = 7

    matrix = [[random.uniform(0, 10) for _ in range(cols)] for _ in range(rows)]

    # Транспонування матриці
    transposed_matrix = np.transpose(matrix)

    # Виведення матриць
    print("Початкова матриця:")
    for row in matrix:
        print(row)

    print("Транспонована матриця:")
    for row in transposed_matrix:
        print(row)

# Завдання 10
# Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть суму добутків елементів кожного стовпця матриці на відповідний елемент вектору. 

def task_10():
    # Розмірність матриці
    rows = 3
    cols = 4

    matrix = [[random.uniform(0, 1) for _ in range(cols)] for _ in range(rows)]
    vector = [random.randint(1, 10) for _ in range(cols)]

    # Знаходження суми добутків
    result = [sum(matrix[i][j] * vector[j] for i in range(rows)) for j in range(cols)]

    # Виведення матриць та вектору
    print("Матриця:")
    for row in matrix:
        print(row)

    print("Вектор:", vector)
    print("Результат:")
    print(result)

# Завдання 11
# Створіть матрицю розміром 2x3 та вектор розміром 3, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть суму добутків елементів кожного рядка матриці на відповідний елемент вектору. 

def task_11():
    # Розмірність матриці
    rows = 2
    cols = 3

    matrix = [[random.uniform(0, 1) for _ in range(cols)] for _ in range(rows)]
    vector = [random.uniform(0, 1) for _ in range(cols)]

    # Знаходження суми добутків
    result = [sum(matrix[i][j] * vector[j] for j in range(cols)) for i in range(rows)]

    # Виведення матриць та вектору
    print("Матриця:")
    for row in matrix:
        print(row)

    print("Вектор:", vector)
    print("Результат:")
    print(result)

# Завдання 12
# Створіть дві матриці розміром 5x5, заповніть їх випадковими цілими числами від 1 до 10 та знайдіть різницю всіх елементів другої матриці.

def task_12():
    # Розмірність матриць
    rows = 5
    cols = 5

    matrix1 = [[random.randint(1, 10) for _ in range(cols)] for _ in range(rows)]
    matrix2 = [[random.randint(1, 10) for _ in range(cols)] for _ in range(rows)]

    # Знаходження різниці елементів
    result_matrix = [[matrix2[i][j] - matrix1[i][j] for j in range(cols)] for i in range(rows)]

    # Виведення матриць та результату
    print("Перша матриця:")
    for row in matrix1:
        print(row)

    print("Друга матриця:")
    for row in matrix2:
        print(row)

    print("Різниця матриць:")
    for row in result_matrix:
        print(row)

# Завдання 13
# Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть добуток їх елементів.

def task_13():
    # Розмірність матриць
    rows = 3
    cols = 3

    matrix1 = [[random.randint(1, 10) for _ in range(cols)] for _ in range(rows)]
    matrix2 = [[random.randint(1, 10) for _ in range(cols)] for _ in range(rows)]

    # Знаходження добутку матриць
    result_matrix = [[matrix1[i][j] * matrix2[i][j] for j in range(cols)] for i in range(rows)]

    # Виведення матриць та результату
    print("Перша матриця:")
    for row in matrix1:
        print(row)

    print("Друга матриця:")
    for row in matrix2:
        print(row)

    print("Добуток матриць:")
    for row in result_matrix:
        print(row)

# Завдання 14
# Створіть матрицю розміром 8x8, заповніть її випадковими цілими числами в діапазоні від 1 до 100 та знайдіть суму елементів на побічній діагоналі.

def task_14():
    # Розмірність матриці
    n = 8

    matrix = [[random.randint(1, 100) for _ in range(n)] for _ in range(n)]

    # Знаходження суми на побічній діагоналі
    diagonal_sum = sum(matrix[i][j] for i, j in enumerate(range(n - 1, -1, -1)))

    # Виведення матриці та суми
    print("Матриця:")
    for row in matrix:
        print(row)

    print("Сума елементів на побічній діагоналі:", diagonal_sum)

# Завдання 15
# Створіть матрицю розміром 5x5, заповніть її випадковими дійсними числами в діапазоні від -2 до 2 та знайдіть вектор-стовпчик, що містить найменше за модулем значення елементу у кожному стовпці матриці.

def task_15():
    # Розмірність матриці
    rows = 5
    cols = 5

    matrix = [[random.uniform(-2, 2) for _ in range(cols)] for _ in range(rows)]

    # Знаходження вектора з найменшими за модулем значеннями у кожному стовпці
    min_magnitude_vector = [min(abs(matrix[i][j]) for i in range(rows)) for j in range(cols)]

    # Виведення матриці та вектора
    print("Матриця:")
    for row in matrix:
        print(row)

    print("Вектор з найменшими за модулем значеннями у кожному стовпці:")
    print(min_magnitude_vector)

def choose_task():
    tasks = {
        1: task_1,
        2: task_2,
        3: task_3,
        4: task_4,
        5: task_5,
        6: task_6,
        7: task_7,
        8: task_8,
        9: task_9,
        10: task_10,
        11: task_11,
        12: task_12,
        13: task_13,
        14: task_14,
        15: task_15,
    }

    task_number = int(input("Введіть номер завдання: "))
    selected_task = tasks.get(task_number)

    if selected_task:
        selected_task()
    else:
        print("Некоректний номер завдання. Спробуйте ще раз.")


if __name__ == "__main__":
    choose_task()


"""
Загальний висновок:
Всі завдання виконуються з використанням стандартних бібліотек Python, таких як random для генерації випадкових чисел та numpy для обробки матриць і векторів.

Код додає можливість вибору конкретного завдання користувачем через введення номера завдання з консолі.
"""
```
